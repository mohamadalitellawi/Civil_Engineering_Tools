```python
def get_m1_m2(m_start:float, m_end:float, Pu:float, h:float) -> float:
    m2_minimum = Pu * (15 + 0.03 * h) # 6.6.4.5.4
    if m_start == 0 and m_end == 0:
        m1 = 0
        m2 = m2_minimum
    elif abs(m_start) <= abs(m_end):
        m1 = m_start
        if m_end >= 0:
            m2 = max(m2_minimum,m_end)
        else:
            m2 = min(-1.0 * m2_minimum, m_end)
    else:
        m1 = m_end
        if m_start >= 0:
            m2 = max(m2_minimum,m_start)
        else:
            m2 = min(-1.0 * m2_minimum, m_start)
    Cm = 0.6 - 0.4 * (m1/m2) * -1
    return Cm,m1,m2

result_minor = [get_m1_m2(m_start, m_end, Pu, h) for m_start, m_end, Pu, h in zip(first_df['M2'], last_df['M2'],first_df['P'],first_df['section_width'])]
result_major = [get_m1_m2(m_start, m_end, Pu, h) for m_start, m_end, Pu, h in zip(first_df['M3'], last_df['M3'],first_df['P'],first_df['section_depth'])]
display(result_major)
display(result_minor)
```

```python
def get_Cm(m_start:float, m_end:float) -> float:
    if abs(m_start) <= abs(m_end):
        m1 = m_start
        m2 = m_end
    else:
        m1 = m_end
        m2 = m_start

    if m2 == 0 :
        Cm = 1.0
    else:
        Cm = 0.6 - 0.4 * (m1/m2) * -1 # we flip by -1 to match etabs results
    return Cm

result_minor = [get_Cm(m_start, m_end) for m_start, m_end in zip(first_df['M2'], last_df['M2'])]
result_major = [get_Cm(m_start, m_end) for m_start, m_end in zip(first_df['M3'], last_df['M3'])]
display(result_major)
display(result_minor)
```

```python
mini_df = pd.DataFrame(
    data = {
        "fc_MPa": [20, 25, 30, 28, 40, 45, 32],
        "gamma": [2350, 1700, 2000, 1600, 2300, 2450, 1680],
    }
)


def concrete_elastic_modulus(fc: float, gamma: float) -> float:
    """
    Returns the elastic modulus of concrete based on the f'c and density of the concrete,
    as per CSA A23.3-04 Cl. 8.6.2.2.
    """   
    if 20 <= fc <= 60:
        return 4500 * fc**0.5
    elif 1500 <= gamma <= 2500:
        return (3300 * fc**0.5 + 6900) * (gamma / 2300) ** 1.5
    else:
        return None
    
mini_df["E"] = mini_df.apply(lambda x: concrete_elastic_modulus(x['fc_MPa'], x['gamma']), axis=1)

mini_df
```

```python
def m2_min(Pu:float, h:float) -> float:
    return Pu * (15 + 0.03 * h) # 6.6.4.5.4

def m2_minimum(Pu:float, h:float, m2:float) -> float:
    m_min = Pu * (15 + 0.03 * h) # 6.6.4.5.4
    m_min = max(m_min, abs(m2)) * np.sign(m2)
    return m_min
```
